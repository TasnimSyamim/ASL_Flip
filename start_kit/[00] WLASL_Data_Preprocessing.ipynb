Copy new videos into existing

```python
import os
import shutil

# Define the source (Folder B) and destination (Folder A)
folder_a = "videos"
folder_b = "WLASL2000"

# Ensure folder paths exist
if not os.path.exists(folder_a) or not os.path.exists(folder_b):
    print("One or both folder paths do not exist.")
else:
    # Iterate through files in Folder B
    for file_name in os.listdir(folder_b):
        src_file = os.path.join(folder_b, file_name)
        dst_file = os.path.join(folder_a, file_name)
        
        # Check if the file already exists in Folder A
        if not os.path.exists(dst_file):
            shutil.copy2(src_file, dst_file)  # Copy with metadata
            print(f"Copied: {file_name}")
        else:
            print(f"Skipped (already exists): {file_name}")
```

Copy into classes

```python
import os
import json
import shutil

# Load dataset
content = json.load(open('WLASL_v0.3.json'))
source_dir = "videos"  # Directory containing the original videos
target_dir = "WLASL-2000"  # Directory to store preprocessed videos

# Ensure target directory exists
os.makedirs(target_dir, exist_ok=True)

i = 0
for entry in content:
    gloss = entry['gloss']  # Gloss name
    instances = entry['instances']
    gloss_dir = os.path.join(target_dir, gloss)
    os.makedirs(gloss_dir, exist_ok=True)  # Create gloss directory
    
    for inst in instances:
        video_id = inst['video_id']
        video_name = f"{video_id}.mp4"  # Expected video file name
        source_path = os.path.join(source_dir, video_name)
        target_path = os.path.join(gloss_dir, video_name)
        
        if os.path.exists(source_path):
            shutil.copy(source_path, target_path)
            print(f"Copied: {video_name} -> {gloss_dir}")
        else:
            print(f"Missing: {video_name}")

print("Processing complete.")
```

Sorting gloss

```python
import os

target_dir = "WLASL-2000"  # Directory containing preprocessed videos

gloss_counts = {}

# Iterate through gloss directories
for gloss in os.listdir(target_dir):
    gloss_path = os.path.join(target_dir, gloss)
    if os.path.isdir(gloss_path):
        video_count = len([f for f in os.listdir(gloss_path) if f.endswith(".mp4")])
        gloss_counts[gloss] = video_count

# Sort by video count in descending order
sorted_gloss_counts = sorted(gloss_counts.items(), key=lambda x: x[1], reverse=True)

# Print results
for gloss, count in sorted_gloss_counts:
    print(f"Gloss: {gloss}, Videos: {count}")
```

Rename files

```python
import os

target_dir = "WLASL-2000"  # Directory containing preprocessed videos

gloss_counts = {}

# Iterate through gloss directories
for gloss in os.listdir(target_dir):
    gloss_path = os.path.join(target_dir, gloss)
    if os.path.isdir(gloss_path):
        video_files = [f for f in os.listdir(gloss_path) if f.endswith(".mp4")]
        video_files.sort()  # Ensure sorted order
        
        # Rename files in numeric order
        for idx, video_file in enumerate(video_files, start=1):
            old_path = os.path.join(gloss_path, video_file)
            new_name = f"{idx}.mp4"
            new_path = os.path.join(gloss_path, new_name)
            os.rename(old_path, new_path)
        
        gloss_counts[gloss] = len(video_files)

# Sort by video count in descending order
sorted_gloss_counts = sorted(gloss_counts.items(), key=lambda x: x[1], reverse=True)

# Print results
for gloss, count in sorted_gloss_counts:
    print(f"Gloss: {gloss}, Videos: {count}")
```

WLASL-100

```python
import os

target_dir = "WLASL-2000"  # Directory containing preprocessed videos
output_file = "WLASL-100.txt"  # Output file to save results

gloss_counts = {}

# Iterate through gloss directories
for gloss in os.listdir(target_dir):
    gloss_path = os.path.join(target_dir, gloss)
    if os.path.isdir(gloss_path):
        video_count = len([f for f in os.listdir(gloss_path) if f.endswith(".mp4")])
        gloss_counts[gloss] = video_count

# Sort by video count in descending order and get top 100
top_100_glosses = sorted(gloss_counts.items(), key=lambda x: x[1], reverse=True)[:100]

# Save top 100 glosses to a file
with open(output_file, "w") as f:
    for gloss, count in top_100_glosses:
        f.write(f"Gloss: {gloss}, Videos: {count}\n")

print(f"Top 100 glosses saved to {output_file}")
```

Copy top-100

```python
import os
import shutil

# Define paths and file name
source_dir = "WLASL-2000"         # Source directory containing gloss folders
destination_dir = "Latest-WLASL-100"      # New destination directory to copy gloss folders
h2s_file = "WLASL-100.txt"         # File that lists the selected glosses

# Create destination directory if it does not exist
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Read gloss names from h2s-100.txt
selected_glosses = []
with open(h2s_file, "r") as file:
    for line in file:
        line = line.strip()
        if line:
            # Each line is expected to be in the format:
            # "Gloss: <gloss_name>, Samples: <count>"
            parts = line.split(",")
            if parts:
                gloss_part = parts[0].strip()  # e.g., "Gloss: book"
                if gloss_part.startswith("Gloss:"):
                    gloss_name = gloss_part[len("Gloss:"):].strip()
                    selected_glosses.append(gloss_name)

print("Glosses to be copied:", selected_glosses)

# Copy each selected gloss folder from source_dir to destination_dir
for gloss in selected_glosses:
    src_path = os.path.join(source_dir, gloss)
    dst_path = os.path.join(destination_dir, gloss)
    if os.path.isdir(src_path):
        # Copy the entire gloss folder, including all its subdirectories.
        shutil.copytree(src_path, dst_path, dirs_exist_ok=True)
        print(f"Copied '{src_path}' to '{dst_path}'")
    else:
        print(f"Gloss '{gloss}' not found in '{source_dir}'")
```
